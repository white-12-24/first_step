```python
now = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' %(now.tm_year, now.tm_mon, now.tm_mday, \
                                      now.tm_hour, now.tm_min, now.tm_sec)

ft_name = ff+s+'-'+q+'\\'+s+'-'+q+'.txt'
fx_name = ff+s+'-'+q+'\\'+s+'-'+q+'.xlsx'
fc_name = ff+s+'-'+q+'\\'+s+'-'+q+'.csv'
hu = ff + s + '-' + q

os.makedirs(hu)
os.chdir(hu)

def scroll_down(driver):      
    driver.execute_script("window.scrollBy(0,120);")
    time.sleep(1)

def scroll_down2(driver):
  driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
  time.sleep(3)
    
no2 = [] # 번호
nn2 = [] # 언론사명
nt2 = [] # 뉴스제목
wd2 = [] # 게시날짜
ws2 = [] # 뉴스내용
no = 1
pcno = 1

for a in range(1, pcnt+1):
    h1 = driver.page_source
    s1 = BeautifulSoup(h1, 'html.parser')
    c1 = s1.find('div', 'news-wp type-wz no-rev').find_all('div', 'news-item')

    for b in range(0, 100):
        scroll_down(driver)
        time.sleep(1)

        no2.append(str(no) + '\n')
        print('%s 번째 기사 정보를 수집합니다=============================================' %str(no))

        try :
            nn1 = c1[b].find('div', 'info').find('a').get_text().strip()

        except :
            nn = c1[b].find('div', class_='info').find('div')
            
            for span in nn.find_all('span'):
                span.decompose()

            nn1 = nn.get_text(strip=True)

        nn2.append(nn1 + '\n')
        print('1.언론사명: ' + nn1)

        nt1 = c1[b].find('span', 'title-elipsis').get_text()
        nt2.append(nt1 + '\n')
        print('2.뉴스제목: ' + nt1)

        wd1 = c1[b].find('p', 'name').get_text()
        wd2.append(wd1 + '\n')
        print('3.게시날짜: '+ wd1)

        driver.find_element(By.XPATH,'//*[@id="news-results"]/div[%s]/div/div[2]/a' %str(b+1)).click( )
        time.sleep(random.randrange(2,5))

        h2 = driver.page_source
        s2 = BeautifulSoup(h2, 'html.parser')
        c3 = s2.find('div', 'news-view')
        
        ws1 = c3.find('div', 'news-view-content').get_text().replace('\n', '').strip()
        ws2.append(ws1 + '\n')
        print('4.뉴스내용: ' + ws1)
        print('\n')

        try :
            driver.find_element(By.XPATH,'//*[@id="news-detail-modal"]/div/div/div/div[3]/button').click()
            time.sleep(random.randrange(2,5))

        except:
            pass
            
        no += 1

        if no > cnt:
            break
                
    pcno += 1
    
    if pcno > pcnt:
        break
        
    try :
        driver.find_element(By.XPATH,'//*[@id="news-results-tab"]/div[6]/div[2]/div/div/div/div/div[4]/a').click( )
        time.sleep(random.randrange(5,10))
        
    except :
        pass
```

```python
bigk = pd.DataFrame()

bigk['번호'] = no2
bigk['언론사명'] = pd.Series(nn2)
bigk['뉴스제목'] = pd.Series(nt2)
bigk['게시날짜'] = pd.Series(wd2)
bigk['뉴스내용'] = pd.Series(ws2)

bigk.to_csv(fc_name, encoding = "utf-8-sig", index = False)
bigk.to_excel(fx_name, index = False)

print('='*80)
print('작업 저장이 완료되었습니다.')
print('='*80)
print()
```
